```python
from collections import defaultdict
from tika import parser
```

```python
stopWords = [
    'Page',
    'Foreign State of Chargeability',
    'Place of Birth',
    'Immigrant Visa Issuances',
    'By Foreign State',
    'or Place of Birth',
]

def hasState(s):
    for stop in stopWords:
        if s.find(stop) == 0 or '(FY' in s:
            return False
    return True

def parseReport(filename):
    raw = parser.from_file(filename)
    rows = list(filter(None, raw['content'].strip().split('\n')))
    states = defaultdict(lambda: defaultdict(int))
    for row in rows:
        if not hasState(row):
            continue
        parts = row.split(' ')
        visa = parts[-2]
        if 'F1' in visa: # Student visa class
            issuances = int(parts[-1].replace(',', ''))
            name = ' '.join(parts[:-2])
            states[name][visa] = issuances
    return states
```

```python
parseReport('../data/JANUARY 2019 - IV Issuances by FSC and Visa Class.pdf')
```
